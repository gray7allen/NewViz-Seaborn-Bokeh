```python
num_ath = pd.DataFrame(data=dfat['sport'].value_counts())
num_ath
```

```python
num_ath.reset_index(inplace=True)
```

```python
num_ath
```

```python
num_ath.rename(columns={'index':'Sport','sport':'Count'},inplace=True)
num_ath
```

```python
topfour = num_ath[:4]
topfour
```

```python
totath = figure(title='Total Athletes',
               x_axis_label='Sport',
               x_range=['athletics','aquatics','football','rowing'],
               y_axis_label='# of Ath.',
               plot_width=600, plot_height=400
               )

totath.vbar(x=['athletics','aquatics','football','rowing'], top=topfour['Count'], width=.5)
show(totath)
```
